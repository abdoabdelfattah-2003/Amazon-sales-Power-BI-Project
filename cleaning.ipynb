## cleaing data

```python
name = 'mohamed_atef'
upper_name = lambda n: n.replace('_', ' ')
print(upper_name(name))
```

```python
df[['actual_price', 'discounted_price']] = df[['actual_price', 'discounted_price']].apply(lambda x: x.str.replace('‚Çπ', ''))
df[['actual_price', 'discounted_price', 'rating_count']] = df[['actual_price', 'discounted_price', 'rating_count']].apply(lambda x: x.str.replace(',', ''))
df['rating'] = df['rating'].str.replace('|', '1')
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')

df[['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']] = df[['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']] .astype('float')
```

## delete unuseful columns

```python
df = df.drop(columns=['review_content' , 'product_link', 'img_link', 'about_product'], axis=1)
```

```python
df.head(10)
```

## add columns

```python
df['price_aftero_discount'] = df.actual_price - df.discounted_price
df.pop('discount_percentage')
df['discount_percentage'] = df['price_aftero_discount'] /  df['actual_price'] 
```

```python
df.rename(columns={'price_aftero_discount' : 'price_after_discount'}, inplace= True)
```

```python
df[['main_category','sub_category']] = df['category'].str.split('|', n=1, expand=True)
# df['sub_category'] = df['sub_category'].str.split('|', n=1).str[0]
# df[['main_category', 'sub_category']].describe()
```

## fill null values

```python
df['rating_count']= df['rating_count'].fillna(df['rating_count'].mean())
```
